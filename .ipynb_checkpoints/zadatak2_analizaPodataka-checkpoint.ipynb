```python
import psycopg2
import pandas as pd
```

a. Izlistati koliki je broj nekretnina za prodaju, a koliki je broj koji se iznajmljuju

```python


try:
        conn = psycopg2.connect(database='nekretnine_database', user='postgres', password = '1234', host='localhost', port=5432)
        cursor = conn.cursor()
        
        count_select_izdavanje = "select count(*) as broj from nekretnine_info where transakcija = 'Izdavanje';"
        cursor.execute(count_select_izdavanje)
        
        records = cursor.fetchone()[0]
        print("Broj nekretnina za izdavanje:", records)
        
        count_select_prodaja = "select count(*) as broj from nekretnine_info where transakcija = 'Prodaja'"
        cursor.execute(count_select_prodaja)
        
        records=cursor.fetchone()[0]
        print("Broj nekretnina za prodaju:", records)
except (Exception, psycopg2.Error) as error:
        print("Neuspesno citanje iz baze", error)
finally:
        if conn:
            cursor.close()
            conn.close()
            
```

b. Izlistati koliko nekretnina se prodaje u svakom od gradova (izlistati sve gradove, obuhvatiti i kuce i stanove)

```python
try:
        conn = psycopg2.connect(database='nekretnine_database', user='postgres', password = '1234', host='localhost', port=5432)
        cursor = conn.cursor()
        
        count_select_grad_brojNekretnina = "select grad, count(*) as broj from nekretnine_info group by grad;"
        cursor.execute(count_select_grad_brojNekretnina)
        
        records = cursor.fetchall();
        for record in records:
            print("Grad:",record[0])
            print("Broj nekretnina:", record[1])
            print("-----------------------------")
        
except (Exception, psycopg2.Error) as error:
        print("Neuspesno citanje iz baze", error)
finally:
        if conn:
            cursor.close()
            conn.close()
```

c. Izlistati koliko uknjizenih, a koliko neuknjizenih kuca, a koliko stanova

```python
try:
        conn = psycopg2.connect(database='nekretnine_database', user='postgres', password = '1234', host='localhost', port=5432)
        cursor = conn.cursor()
        
        count_select_uknjizeno_stanovi = "select uknjizeno, count(*) as broj_stanova from nekretnine_info where kategorija = 'stan' group by uknjizeno;"
        cursor.execute(count_select_uknjizeno_stanovi)
        
        records = cursor.fetchall()
        for record in records:
            uknjizeno = "Ne"
            if record[0] == "Da":
                uknjizeno = record[0]
            print("Uknjizeno: ", uknjizeno)
            print("Broj stanova:", record[1])
            
        print('-----------------------------')
            
        count_select_uknjizeno_kuce ="select uknjizeno, count(*) as broj_kuca from nekretnine_info where kategorija = 'kuća' group by uknjizeno;"
        cursor.execute(count_select_uknjizeno_kuce)
        
        records = cursor.fetchall()
        for record in records:
            uknjizeno = "Ne"
            if record[0] == "Da":
                uknjizeno = record[0]
            print("Uknjizeno: ", uknjizeno)
            print("Broj kuća", record[1])
        
    
except (Exception, psycopg2.Error) as error:
        print("Neuspesno citanje iz baze", error)
finally:
        if conn:
            cursor.close()
            conn.close()
```

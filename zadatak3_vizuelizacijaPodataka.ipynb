```python
import psycopg2
import pandas as pd
```

```python
import matplotlib.pyplot as plt

def draw_rank_correlation(x_rank, y_rank, x_label, y_label, how="bar"):                                                    
    plt.figure(num=1, figsize=[20, 5])

    labels = x_rank
    plt.xlabel(x_label)
    
    plt.ylabel(y_label)                                                                                                          
    plt.yscale('linear')                                                                                                                
    plt.ylim(1, max(y_rank) * 1.1)                                                                                                             
    
    if how == "plot":
        plt.plot(x_rank, y_rank, marker='.')
    else:
        plt.bar(x_rank, y_rank, tick_label=x_rank, width = 0.5, color = ['red', 'green'])
    plt.show()
```

a. 10 najzastupljenijih delova Beograda koji imaju najveci broj nekretnina u ponudi (i u sekciji za prodaju, i u sekciji za iznajmljivanje i zbirno)

```python
try:
        conn = psycopg2.connect(database='nekretnine_database', user='postgres', password = '1234', host='localhost', port=5432)
        cursor = conn.cursor()
        
        #prodaja
        print("Top 10 najzastupljenijih delova Beograda koji imaju najveci broj nekretnina u ponudi za prodaju")
        top_10_deloviBeograda_Prodaja = "select lokacija, count(*) as broj_nekretnina from nekretnine_info where grad = 'Beograd' and lokacija is not null and transakcija = 'Prodaja' group by lokacija order by broj_nekretnina desc limit 10;"
        cursor.execute(top_10_deloviBeograda_Prodaja)
        records = cursor.fetchall()
        col_names = []
        for elt in cursor.description:
            col_names.append(elt[0])
        df_prodaja = pd.DataFrame(records, columns=col_names)
        print(df_prodaja)
        draw_rank_correlation(df_prodaja['lokacija'], df_prodaja['broj_nekretnina'], 'lokacija', 'broj_nekretnina', how='bar')
        
        #izdavanje
        print("Top 10 najzastupljenijih delova Beograda koji imaju najveci broj nekretnina u ponudi za prodaju")
        top_10_deloviBeograda_Izdavanje = "select lokacija, count(*) as broj_nekretnina from nekretnine_info where grad = 'Beograd' and lokacija is not null and transakcija = 'Izdavanje' group by lokacija order by broj_nekretnina desc limit 10;"
        cursor.execute(top_10_deloviBeograda_Izdavanje)
        records = cursor.fetchall()
        col_names = []
        for elt in cursor.description:
            col_names.append(elt[0])
        df_izdavanje = pd.DataFrame(records, columns=col_names)
        print(df_izdavanje)
        draw_rank_correlation(df_izdavanje['lokacija'], df_izdavanje['broj_nekretnina'], 'lokacija', 'broj_nekretnina', how='bar')
        
        #kombinacija
        print("Top 10 najzastupljenijih delova Beograda koji imaju najveci broj nekretnina u ponudi za prodaju")
        top_10_deloviBeograda = "select lokacija, count(*) as broj_nekretnina from nekretnine_info where grad = 'Beograd' and lokacija is not null group by lokacija order by broj_nekretnina desc limit 10;"
        cursor.execute(top_10_deloviBeograda)
        records = cursor.fetchall()
        col_names = []
        for elt in cursor.description:
            col_names.append(elt[0])
        df = pd.DataFrame(records, columns=col_names)
        print(df)
        draw_rank_correlation(df['lokacija'], df['broj_nekretnina'], 'lokacija', 'broj_nekretnina', how='bar')
        
except (Exception, psycopg2.Error) as error:
        print("Neuspesno citanje iz baze", error)
finally:
        if conn:
            cursor.close()
            conn.close()
```

d. Broj (i procentualni odnos) nekretnina koje se prodaju i nekretnina koje se iznajmljuju, za prvih 5 gradova sa najvecim brojem nekretnina (za svaki grad posebno prikazati grafikon BROJ_ZA_PRODAJU:BROJ_ZA_IZNAJMLJIVANJE)

```python
try:
        conn = psycopg2.connect(database='nekretnine_database', user='postgres', password = '1234', host='localhost', port=5432)
        cursor = conn.cursor()
        
        #izdvoj top 5 grada
        print("Top 5 grada sa najvecim brojem nekretnina")
        top_5_grad = "select grad, count(*) as broj from nekretnine_info group by grad order by broj desc limit 5 ;"
        cursor.execute(top_5_grad)
        records = cursor.fetchall()
        col_names = []
        for elt in cursor.description:
            col_names.append(elt[0])
        df = pd.DataFrame(records, columns=col_names)
        for grad in df["grad"]:
            print("Racunanje odnosa broja nekretnina za prodaju i nekretnina za iznajmiti za grad", grad, "...")
            grad_upit = "select transakcija, count(*) as broj from nekretnine_info where grad = '" + grad + "' group by transakcija;"
            cursor.execute(grad_upit)
            records = cursor.fetchall()
            col_names = []
            for elt in cursor.description:
                col_names.append(elt[0])
            df = pd.DataFrame(records, columns=col_names)
            print(df)
            draw_rank_correlation(df['transakcija'], df['broj'], 'transakcija', 'broj_nekretnina', how='bar')
            
        
except (Exception, psycopg2.Error) as error:
        print("Neuspesno citanje iz baze", error)
finally:
        if conn:
            cursor.close()
            conn.close()
```

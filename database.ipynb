```python
#Selektovani svi podaci iz tabele nekretnine_info. Ukupan broj podataka je 23767 
import psycopg2
import pandas as pd

try:
        conn = psycopg2.connect(database='nekretnine_database', user='postgres', password = '1234', host='localhost', port=5432)
        cursor = conn.cursor()
        
        select_database_query = "select * from nekretnine_info"
        cursor.execute(select_database_query)
        
        print("Selektovani svi podaci iz tabele nekretnine_info")
        records = cursor.fetchall()
        print(len(records))
except (Exception, psycopg2.Error) as error:
        print("Neuspesno citanje iz baze", error)
finally:
        if conn:
            cursor.close()
            conn.close()
            print("POSTGRESQL konekcija zatvorena")
            

#write data into csv
df = pd.DataFrame(records)
print(df)
df.to_csv("nekretnine_info.csv")
```
